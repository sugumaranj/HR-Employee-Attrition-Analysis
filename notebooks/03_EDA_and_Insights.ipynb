# Attrition Rate Function

To avoid repeating calculations, a reusable function is created.

The function calculates attrition percentages for any categorical column.

```python
# Function to calculate attrition rates

def attrition_rate(df, column):

    result = (
        pd.crosstab(
            df[column],
            df["Attrition"],
            normalize="index"
        ) * 100
    )

    return result.round(2)
```

```python
#test call to the function
attrition_rate(df, "Department")
```

### Observation

The function returns attrition percentages
for every category within a selected feature.

This function will be reused throughout the project.

# EDA Summary & Conclusion

## Key Insights

- **Overall Attrition:** The company has an employee attrition rate of **16.12%**.
- **Highest Risk Department:** The **Sales Department** shows the highest attrition rate (**20.63%**) among all departments.
- **Highest Risk Role:** **Sales Representatives** exhibit the highest attrition rate (**39.76%**) among all job roles.
- **Burnout Factors:** Employees working **Overtime** have an attrition rate of **30.53%**, while employees who travel **Frequently** for business show an attrition rate of **24.91%**.
- **Employee Well-being:** Lower **Work-Life Balance** and **Job Satisfaction** levels are associated with higher attrition rates.
- **Compensation & Experience:** Employees who left earned an average monthly income of **₹4,787.09**, compared to **₹6,832.74** for employees who stayed.
- **Experience & Tenure:** Employees who left had an average of **8.24 years** of work experience and **5.13 years** at the company, compared to **11.86 years** and **7.37 years** respectively for retained employees.

---

## Conclusion & Recommendations

The analysis indicates that employee attrition is strongly associated with overtime work, frequent business travel, lower satisfaction levels, lower compensation, and shorter tenure. Sales-related roles, particularly Sales Representatives, exhibit the highest turnover rates.

### Recommendations

1. **Prioritize Sales Retention:** Develop targeted retention programs for Sales Representatives and employees in the Sales department.
2. **Reduce Burnout:** Monitor overtime workloads and explore alternatives to frequent business travel where possible.
3. **Improve Employee Well-being:** Strengthen work-life balance initiatives and employee satisfaction programs.
4. **Support Early-Career Employees:** Review compensation structures, onboarding processes, and career development opportunities for less experienced employees.

# Correlation Analysis

```python
df["AttritionNumeric"] = df["Attrition"].map({
    "No": 0,
    "Yes": 1
})
```

```python
numerical_df = df.select_dtypes(include="number")

attrition_corr = (
    numerical_df
    .corr()["AttritionNumeric"]
    .drop("AttritionNumeric")
    .sort_values(ascending=False)
)

print(attrition_corr)
```

### Correlation Analysis Summary

Correlation analysis revealed that employee attrition is most negatively associated with Total Working Years (-0.171), Job Level (-0.169), Years in Current Role (-0.161), Monthly Income (-0.160), and Age (-0.159).

These findings indicate that employees who are more experienced, occupy higher job positions, earn higher salaries, and have longer tenure are generally less likely to leave the organization.

Distance From Home (0.078) showed the strongest positive correlation with attrition, suggesting that employees living farther from the workplace may have a slightly higher tendency to leave.
