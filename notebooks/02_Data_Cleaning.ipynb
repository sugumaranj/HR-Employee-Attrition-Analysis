# Stage 2: Data Cleaning

## Objective

The objective of this stage is to assess and improve the quality of the dataset before performing any analysis.

Data cleaning helps ensure that the insights generated later are accurate and reliable.

# Missing Value Analysis

Missing values can affect statistical calculations and lead to incorrect conclusions.

We first check whether any column contains missing values.

```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the HR Attrition dataset
try:
    df = pd.read_csv("../data/employee_attrition.csv")
except FileNotFoundError:
    print("File not found")
except Exception:
    print("An Error occured")
else:
    print("File loaded successfully")
```

```python
# Check missing values in each column

df.isnull().sum()
```

### Observation

The output displays the number of missing values in each column.

If all values are zero, the dataset contains no missing values.

```python
# Total missing values in the dataset

df.isnull().sum().sum()
```

### Observation

The total number of missing values in the dataset is 0.

# Duplicate Record Analysis

Duplicate records can distort analysis by counting the same employee multiple times.

We check whether duplicate rows exist.

```python
# Count duplicate rows

df.duplicated().sum()
```

### Observation

The dataset contains 0 duplicate records.

# Unique Value Analysis

Columns containing only one unique value do not provide useful information for analysis.

```python
# Count unique values in every column

df.nunique()
```

### Observation

The following columns contain only one unique value:

- EmployeeCount
- Over18
- StandardHours

These columns provide no analytical value and can be removed.

# Removing Unnecessary Columns

Columns with no variability do not contribute to analysis.

EmployeeNumber is an identifier and is not useful for finding business insights.

```python
# Create cleaned dataset

cleaned_df = df.drop(
    columns=[
        "EmployeeCount",
        "Over18",
        "StandardHours",
        "EmployeeNumber"
    ]
)
```

```python
# Verify remaining columns

cleaned_df.shape
```

### Observation

The cleaned dataset contains fewer columns after removing constant-value and identifier columns.

# Validation

After cleaning, we verify that the dataset is ready for analysis.

```python
cleaned_df.info()
```

```python
cleaned_df.head()
```

```python
cleaned_df.tail(10)
```

### Observation

The cleaned dataset appears consistent and ready for exploratory data analysis.

```python
# Save cleaned dataset
try:
    cleaned_df.to_csv(
        "../data/CleanedDataset/employee_cleaned.csv",
        index=False
    )
except Exception:
    print("An Error occured")
else:
    print("Cleaned dataset saved successfully")
```

### Observation

The cleaned dataset has been saved successfully and will be used for all future analyses.

# Conclusion

Data cleaning was performed successfully.

The dataset was checked for:

- Missing values
- Duplicate records
- Constant-value columns

The following columns were removed:

- EmployeeCount
- Over18
- StandardHours
- EmployeeNumber

The cleaned dataset is now ready for Exploratory Data Analysis (EDA).
