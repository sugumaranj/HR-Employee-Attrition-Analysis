# Stage 4: Data Visualization

## Objective

The objective of this stage is to visually represent the findings from the exploratory data analysis.

Visualizations help identify patterns, trends, and relationships more effectively than raw numerical tables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("../data/CleanedDataset/employee_cleaned.csv")
```

```python
import os

os.makedirs("../charts", exist_ok=True)
```

# Attrition Distribution

### Business Question

What percentage of employees left the company?

```python
attrition_counts = df["Attrition"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    attrition_counts,
    labels=attrition_counts.index,
    autopct="%1.1f%%",
    colors=["#4E79A7", "#E15759"]
)

plt.title("Employee Attrition Distribution")

plt.savefig(
    "../charts/attrition_distribution.png",
    bbox_inches="tight"
)

plt.show()
```

### Observation

Most employees remained in the company, while a smaller proportion left.

# Department vs Attrition Rate

```python
department_rate = (
    pd.crosstab(
        df["Department"],
        df["Attrition"],
        normalize="index"
    ) * 100
)

department_rate["Yes"].sort_values(
    ascending=False
).plot(
    kind="bar",
    figsize=(8,5),
    color="#4E79A7"
)

plt.title("Department Attrition Rate")
plt.ylabel("Attrition Percentage")

plt.savefig(
    "../charts/department_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

### Observation

Sales department exhibits the highest attrition rate.

# Overtime vs Attrition

```python
overtime_rate = (
    pd.crosstab(
        df["OverTime"],
        df["Attrition"],
        normalize="index"
    ) * 100
)

overtime_rate["Yes"].plot(
    kind="bar",
    figsize=(6,4),
    color="#4E79A7"
)

plt.title("Overtime vs Attrition")
plt.ylabel("Attrition Percentage")

plt.savefig(
    "../charts/overtime_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

### Observation

Employees working overtime show significantly higher attrition.

# Job Satisfaction vs Attrition

```python
jobsat_rate = (
    pd.crosstab(
        df["JobSatisfaction"],
        df["Attrition"],
        normalize="index"
    ) * 100
)

jobsat_rate["Yes"].plot(
    kind="bar",
    figsize=(6,4),
    color="#4E79A7"
)

plt.title("Job Satisfaction vs Attrition")
plt.ylabel("Attrition Percentage")

plt.savefig(
    "../charts/job_satisfaction_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

# Work-Life Balance vs Attrition

```python
wlb_rate = (
    pd.crosstab(
        df["WorkLifeBalance"],
        df["Attrition"],
        normalize="index"
    ) * 100
)

wlb_rate["Yes"].plot(
    kind="bar",
    figsize=(6,4),
    color="#4E79A7"
)

plt.title("Work-Life Balance vs Attrition")
plt.ylabel("Attrition Percentage")

plt.savefig(
    "../charts/worklifebalance_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

# Business Travel vs Attrition

```python
travel_rate = (
    pd.crosstab(
        df["BusinessTravel"],
        df["Attrition"],
        normalize="index"
    ) * 100
)

travel_rate["Yes"].sort_values(
    ascending=False
).plot(
    kind="bar",
    figsize=(8,5),
    color="#4E79A7"
)

plt.title("Business Travel vs Attrition")
plt.ylabel("Attrition Percentage")

plt.savefig(
    "../charts/business_travel_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

# Job Role vs Attrition

```python
jobrole_rate = (
    pd.crosstab(
        df["JobRole"],
        df["Attrition"],
        normalize="index"
    ) * 100
)

jobrole_rate["Yes"].sort_values(
    ascending=True
).plot(
    kind="barh",
    figsize=(10,6),
    color="#4E79A7"
)

plt.title("Job Role vs Attrition")
plt.xlabel("Attrition Percentage")

plt.savefig(
    "../charts/jobrole_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

# Monthly Income vs Attrition

### Business Question

Do employees who leave the company earn less than employees who stay?

```python
salary_data = (
    df.groupby("Attrition")["MonthlyIncome"]
      .mean()
      .round(2)
)

salary_data.plot(
    kind="bar",
    figsize=(6,4),
    color="#4E79A7"
)

plt.title("Average Monthly Income by Attrition")
plt.ylabel("Average Monthly Income")

plt.savefig(
    "../charts/monthly_income_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

### Observation

Employees who left the company earned a lower average monthly income than employees who stayed.

# Total Working Experience vs Attrition

### Business Question

Do employees with less work experience leave more frequently?

```python
experience_data = (
    df.groupby("Attrition")["TotalWorkingYears"]
      .mean()
      .round(2)
)

experience_data.plot(
    kind="bar",
    figsize=(6,4),
    color="#4E79A7"
)

plt.title("Average Total Working Years by Attrition")
plt.ylabel("Average Years of Experience")

plt.savefig(
    "../charts/experience_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

### Observation

Employees who left the company had fewer years of work experience on average compared to employees who stayed.

# Years at Company vs Attrition

```python
tenure_data = (
    df.groupby("Attrition")["YearsAtCompany"]
      .mean()
      .round(2)
)

tenure_data.plot(
    kind="bar",
    figsize=(6,4),
    color="#4E79A7"
)

plt.title("Average Years at Company by Attrition")
plt.ylabel("Average Years at Company")

plt.savefig(
    "../charts/tenure_attrition.png",
    bbox_inches="tight"
)

plt.show()
```

### Observation

Employees who left the company had shorter average tenure compared to retained employees.

# Visualization Summary

- Overtime, frequent business travel, and poor work-life balance are strongly associated with higher attrition.
- Sales Representatives and employees in the Sales department show the highest turnover rates.
- Lower job satisfaction are linked to increased attrition.
- Employees who left generally had lower salaries, less experience, and shorter tenure.

### Conclusion

The visualizations indicate that workload, employee satisfaction, compensation, and experience are key factors associated with employee attrition. Addressing these areas can help improve employee retention and reduce turnover.
